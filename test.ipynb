```python
import imageio
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from skimage.transform import resize
from IPython.display import HTML

def display_video(video):
    fig = plt.figure(figsize=(3,3))  #Display size specification

    mov = []
    for i in range(len(video)):  #Append videos one by one to mov
        img = plt.imshow(video[i], animated=True)
        plt.axis('off')
        mov.append([img])

    #Animation creation
    anime = animation.ArtistAnimation(fig, mov, interval=50, repeat_delay=1000)

    plt.close()
    return anime
```

```python
from deep_sort_realtime.deepsort_tracker import DeepSort
from ultralytics import YOLO
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
# Path to the video file
video_path = "/content/drive/MyDrive/aiproject/sample.mp4"

# Confidence threshold for detection
conf_threshold = 0.5

# Class to track (e.g., class ID 1)
tracking_class = 1

# Initialize the tracker with a maximum age of 5
tracker = DeepSort(max_age=5)

# Load the YOLO model
model = YOLO("/content/best.pt")

# Define class names
class_names = ["ball", "net"]

# Generate random colors for each class
colors = np.random.randint(0, 255, size=(len(class_names), 3))

# List to hold track information
tracks = []

# Open the video file
cap = cv2.VideoCapture(video_path)

while True:
    # Read a frame from the video
    ret, frame = cap.read()

    if not ret:
        break  # Exit loop if no more frames

    # Get predictions from the model
    results = model(frame)

    detect = []

    for detect_object in results[0].boxes:
        label, confidence, bbox = detect_object.cls, detect_object.conf, detect_object.xyxy[0]
        x1, y1, x2, y2 = map(int, bbox)
        class_id = int(label)

        if tracking_class is None:
            if confidence < conf_threshold:
                continue
        else:
            if class_id != tracking_class or confidence < conf_threshold:
                continue

        # Append detection as [x1, y1, width, height], confidence, class_id
        detect.append([[x1, y1, x2-x1, y2-y1], confidence, class_id])

    # Update tracker with the detected objects
    tracks = tracker.update_tracks(detect, frame=frame)

    for track in tracks:
        if track.is_confirmed():
            track_id = track.track_id

            # Get bounding box and class for the track
            ltrb = track.to_ltrb()
            class_id = track.get_det_class()
            x1, y1, x2, y2 = map(int, ltrb)
            color = colors[class_id]
            B, G, R = map(int, color)

            # Generate the label text
            label = "{}-{}".format(class_names[class_id], track_id)

            # Draw bounding box and label on the frame
            cv2.rectangle(frame, (x1, y1), (x2, y2), (B, G, R), 2)
            cv2.rectangle(frame, (x1 - 1, y1 - 20), (x1 + len(label) * 12, y1), (B, G, R), -1)
            cv2.putText(frame, label, (x1 + 5, y1 - 8), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2)

    # Display the frame
    HTML(display_video(frame).to_html5_video())  #Inline video display in HTML5
    # Break loop if 'q' is pressed
    if cv2.waitKey(1) == ord("q"):
        break

# Release video capture and close windows
cap.release()
cv2.destroyAllWindows()
```
